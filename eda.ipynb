```python
# load libraries for exploring data

import pandas as pd
```

```python
# load data

data = pd.read_csv('items.csv')
```

```python
data
```

```python
# convert title to lowercase

data['title'] = data['title'].str.lower()
```

```python
# !pip install sentence-transformers

# create text embedding version of title column items

from sentence_transformers import SentenceTransformer

model = SentenceTransformer('paraphrase-multilingual-mpnet-base-v2')

# data['title_embedding'] = data['title'].apply(lambda x: model.encode(x))

model.encode(data['title'][0])
```

```python
# create a sepaarate dataframe for the embeddings, where each row is an item and the columns are the embeddings, the id of the item should be the index

embeddings = pd.DataFrame(data['title'].apply(lambda x: model.encode(x)).tolist(), index=data.index)
```

```python
# save the embeddings to a csv file

embeddings.to_csv('title_embeddings.csv')
```
